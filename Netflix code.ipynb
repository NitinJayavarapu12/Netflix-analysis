```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('netflix_titles.csv')
df = pd.DataFrame(data)

df.head()
```

```python
df.head()
```

### Data Preparation

```python
# Date Conversion
df["date_added"] = pd.to_datetime(df["date_added"])

# Duration - hot encoding
df["is_movie"] = df["duration"].fillna("").str.contains("min").astype(int)

# Movie Duration
df["movie_duration"] = df["duration"].fillna("").str.replace(" min", "").where(df["is_movie"] == 1, 0).astype(int)

# Season Count
df["season_count"] = df["duration"].fillna("").replace({" Season": "", " Seasons": ""}).where(df["is_movie"] == 0, 0).astype(int)
```
